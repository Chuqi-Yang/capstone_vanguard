```
total_df['MktCapSize'] = 'NaN'
print(total_df.head())
```

```
total_df.MktCapSize[total_df.MktCap >= 2e11] = 'MegaCap'
total_df[total_df['MktCapSize'] == 'MegaCap']
```

```
total_df.MktCapSize[(total_df.MktCap >= 1e10) & (total_df.MktCap < 2e11) ] = 'LargeCap'
total_df[total_df['MktCapSize'] == 'LargeCap']
```

```
total_df.MktCapSize[(total_df.MktCap >= 2e9) & (total_df.MktCap < 1e10) ] = 'MidCap'
total_df[total_df['MktCapSize'] == 'MidCap']
```

```
total_df.MktCapSize[(total_df.MktCap >= 3e8) & (total_df.MktCap < 2e9) ] = 'SmallCap'
total_df[total_df['MktCapSize'] == 'SmallCap']
```

```
total_df.MktCapSize[(total_df.MktCap >= 5e7) & (total_df.MktCap < 3e8) ] = 'MicroCap'
total_df[total_df['MktCapSize'] == 'MicroCap']
```

```
total_df.MktCapSize[total_df.MktCap < 5e7] = 'NanoCap'
total_df[total_df['MktCapSize'] == 'NanoCap']
```
