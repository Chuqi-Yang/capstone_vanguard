```
def get_time_series_features_for_all_tps(dfs):
  new_dfs = []
  for df in dfs:
    new_df = df.loc[(df.HOLDDATE == '2018-12-31') | (df.HOLDDATE == '2020-03-31') | (df.HOLDDATE == '2020-06-30')]
    new_dfs.append(new_df)
  return new_dfs
```

```
# get all the five features on the given time point
dfs = [df_turnover, df_mktcap, df_num_of_instr, df_concentration, df_total_assets, df_inv_style]
df_turnover_ts, df_mktcap_ts, df_num_of_instr_ts, df_concentration_ts, df_total_assets_ts = get_time_series_features_for_all_tps(dfs[:-1])
```

```
max_scaler = lambda x : x / np.max(x)
```

```
# Normalize turnover rate
df_turnover_ts.iloc[:, -1] = np.log(df_turnover_ts.iloc[:, -1] + 1)
y = df_turnover_ts[['Turnover']].apply(max_scaler)
df_turnover_ts = df_turnover_ts.drop(columns='Turnover')
df_turnover_ts['Turnover_Max'] = y
df_turnover_ts
```

```
# Normalize number of instruments
df_num_of_instr_ts.iloc[:, -1] = np.log(df_num_of_instr_ts.iloc[:, -1])
y = df_num_of_instr_ts[['NumberOfInstr']].apply(max_scaler)
df_num_of_instr_ts = df_num_of_instr_ts.drop(columns='NumberOfInstr')
df_num_of_instr_ts['NumberOfInstr_Max'] = y
df_num_of_instr_ts
```

```
# Normalize total assets
df_total_assets_ts.iloc[:, -1] = np.log(df_total_assets_ts.iloc[:, -1])
y = df_total_assets_ts[['TotalAssets']].apply(max_scaler)
df_total_assets_ts = df_total_assets_ts.drop(columns='TotalAssets')
df_total_assets_ts['TotalAssets_Max'] = y
df_total_assets_ts
```

```
# 'dfs_ts' includes 5 time series features on all time points (not including 'inv style').
dfs_ts = [df_turnover_ts, df_mktcap_ts, df_num_of_instr_ts, df_concentration_ts, df_total_assets_ts]

# 'df_x_ts' is the merged data frame of all the 5 time series features
df_x_ts = reduce(lambda df_left, df_right: pd.merge(df_left, df_right, 
                                                    on=['LGCYINVESTORID', 'HOLDDATE'], 
                                                    how='outer').fillna(0), 
                 dfs_ts)
```

```
# Drop some of the time series features that have correlations inbetween
df_x_ts = df_x_ts.drop(columns=['PctOfSmallCap', 'PctOfNanoCap'])

# Combine time series features with the static feature 'inv style'
# Also drop some small features with all zeros or have strong correlations inbetween
df_x_ts_merged = df_x_ts.merge(df_inv_style, on='LGCYINVESTORID', how='outer').fillna(0).drop(columns=['InvStyle_Technicals', 'InvStyle_Growth','InvStyle_Aggres. Gr.'])
df_x_ts_merged
```

## Computing y in one significant industry on one specific time point 

For each investor $i$ on time point $t$, when we are considering about one specific industry $D$, the dependent variable $y_{i}$ follows the equation:
\begin{align*}
\Delta_{it} &= \frac{1}{n}\sum_{j \in \text D} \Delta_{ijt}\\
y_i &= |\Delta_{it} - \bar{\Delta_{t}}|
\end{align*}

$n$ = # of instruments that investor $i$ invested in industry $D$





```
df_delta_weighted = pd.read_csv('./data for trend/225 investors/delta_with_weight.csv').drop(columns=['Unnamed: 0'])
df_delta_weighted
```

```
df_instr_indus = pd.read_csv('./full database/instrument_industry_match_title.csv')
df_instr_indus
```

```
df_delta_weighted_with_industry = pd.merge(df_delta_weighted, df_instr_indus, on='INSTRID', how='inner')
df_delta_weighted_with_industry
```

```
# Specify time point and industry
df_delta_181231_55 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2018-12-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 55)]
df_delta_200630_55 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-06-30') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 55)]
df_delta_200331_59 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-03-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 59)]
df_delta_200331_51 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-03-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 51)]
df_delta_200331_50 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-03-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 50)]
df_delta_200331_60 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-03-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 60)]
df_delta_200331_53 = df_delta_weighted_with_industry[(df_delta_weighted_with_industry['HOLDDATE'] == '2020-03-31') & (df_delta_weighted_with_industry['HierarchicalId_2'] == 53)]
```

```
df_delta_all_tps = pd.concat([df_delta_181231_55, df_delta_200630_55, df_delta_200331_59, df_delta_200331_51, df_delta_200331_50, df_delta_200331_60, df_delta_200331_53], axis=0)
df_delta_all_tps = df_delta_all_tps.drop(columns=['HierarchicalId_x', 'Title_4digits'])
df_delta_all_tps
```

```
df_y_agg_by_inv = pd.pivot_table(df_delta_all_tps, index=['LGCYINVESTORID', 'HOLDDATE', 'HierarchicalId_2'], values='Delta', aggfunc=np.mean).reset_index()
df_y_agg_by_inv
```

```
df_y_tp_agg_by_indus = pd.pivot_table(df_y_agg_by_inv, index=['HOLDDATE', 'HierarchicalId_2'], values='Delta', aggfunc=np.mean).reset_index()
df_y_tp_agg_by_indus = df_y_tp_agg_by_indus.rename(columns={'Delta': 'Mean_Delta'})
```

```
df_y = pd.merge(df_y_agg_by_inv, df_y_tp_agg_by_indus, on=['HOLDDATE', 'HierarchicalId_2'], how='inner')
df_y
```

```
df_y['y'] = abs(df_y.Delta - df_y.Mean_Delta)
df_y = df_y.drop(columns=['Delta', 'Mean_Delta'])
df_y
```

## Regression Analysis using 6 features

```
df_lr = pd.merge(df_x_ts_merged, df_y, on=['LGCYINVESTORID', 'HOLDDATE'], how='inner')
dummies = pd.get_dummies(df_lr.loc[:, 'HierarchicalId_2'], prefix='HierarchicalId' ) 
df_lr = pd.concat([df_lr, dummies], axis = 1).drop(columns='HierarchicalId_2')
df_lr
```
